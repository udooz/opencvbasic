# Annotating Images

Drawing lines, circles, rectangles and add text

```python
import cv2 as cv
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

matplotlib.rcParams['figure.figsize'] = (9.0, 9.0)

%matplotlib inline
```

```python
app_img = cv.imread('./nb3/Apollo_11_Launch.jpg', cv.IMREAD_COLOR)
plt.imshow(app_img[:,:,::-1])
```

### Drawing Line or Rectangle

`
img = cv2.line(img, pt1, pt2, color[, thickness[, lineType[, shift]]])
img = cv2.rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]])
`
`
pt1: (x1, y1), pt2: (x2, y2)

color: Color of the line which will be drawn

thickness: Integer specifying the line thickness. Default value is 1.

lineType: Type of the line. Default value is 8 which stands for an 8-connected line. Usually, cv2.LINE_AA (antialiased or smooth line) is used for the lineType.

### Drawing Circle
`
img = cv2.circle(img, center, radius, color[, thickness[, lineType[, shift]]])
`

### Adding Text
`img = cv2.putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]])`

org - bottom-left corner of the text string in the image
color - font color

```python
app_img_line = app_img.copy()
cv.line(app_img_line, (200, 100), (400, 100), (0, 255, 255), thickness=5, lineType=cv.LINE_AA)
plt.imshow(app_img_line[:,:,::-1])
```

```python
app_img_circle = app_img.copy()
cv.circle(app_img_circle, (900, 500), 100, (0, 0, 255), thickness=5, lineType=cv.LINE_AA)
plt.imshow(app_img_circle[:,:,::-1])
```

```python
app_img_rect = app_img.copy()
# if thickness is -5, filled rectangle
cv.rectangle(app_img_rect, (500, 100), (700, 600), (255, 0, 255), thickness=5, lineType=cv.LINE_AA)
plt.imshow(app_img_rect[:,:,::-1])
```

```python
app_img_text = app_img.copy()
text = "Apollo 11 Saturn V Launch, July 16, 1969"
fontScale = 2.3
fontFace = cv.FONT_HERSHEY_PLAIN
fontColor = (0, 255, 0)
fontThickness = 2

cv.putText(app_img_text, text, (200, 700), fontFace, fontScale, fontColor, fontThickness, cv.LINE_AA)

# Display the image
plt.imshow(app_img_text[:, :, ::-1])
```
